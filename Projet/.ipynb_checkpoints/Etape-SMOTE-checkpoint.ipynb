```python
# üì¶ Imports
import pandas as pd
import numpy as np
import re
import emoji
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import scipy.sparse

# ‚öôÔ∏è 1. Fusion des classes (CLAIM/REF vs CONTEXT)
claim_ref['label'] = 1
context_only['label'] = 0
df_text = pd.concat([claim_ref, context_only], ignore_index=True)
df_text['label'] = df_text['label'].astype(int)

# üßº 2. Nettoyage du texte (emojis) et d√©tection de patterns
def convert_emojis(text):
    return emoji.demojize(str(text))

df_text['tweet_text'] = df_text['text'].apply(convert_emojis)

df_text['has_url'] = df_text['text'].str.contains(r'http[s]?://', regex=True)
df_text['has_mention'] = df_text['text'].str.contains(r'@\w+', regex=True)
df_text['has_hashtag'] = df_text['text'].str.contains(r'#\w+', regex=True)
df_text['has_emoji'] = df_text['tweet_text'].str.contains(r':[^:\s]+:', regex=True)

# üß† 3. Vectorisation TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, min_df=3, max_df=0.95)
X_tfidf = vectorizer.fit_transform(df_text['tweet_text'])

# üßÆ 4. Filtrage des tokens dans TF-IDF
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer.get_feature_names_out())

def get_token_category(token):
    if token.isdigit():
        return 'numeric'
    if re.match(r'^\d+(am|pm|s|h|min|sec|hour|years)?$', token.lower()):
        return 'timestamp'
    if re.match(r'^[a-zA-Z]+$', token):
        return 'valid'
    return 'other'

token_categories = {token: get_token_category(token) for token in tfidf_df.columns}

valid_tokens = [token for token, cat in token_categories.items() if cat == 'valid']
numeric_tokens = [token for token, cat in token_categories.items() if cat == 'numeric']
timestamp_tokens = [token for token, cat in token_categories.items() if cat == 'timestamp']

# Colonnes agr√©g√©es
tfidf_df['numeric'] = tfidf_df[numeric_tokens].sum(axis=1) if numeric_tokens else 0
tfidf_df['timestamp'] = tfidf_df[timestamp_tokens].sum(axis=1) if timestamp_tokens else 0

# üßº 5. Nettoyage de la matrice TF-IDF
tfidf_cleaned = tfidf_df[valid_tokens + ['numeric', 'timestamp']]

print("üìÑ Aper√ßu de la matrice TF-IDF nettoy√©e :")
display(tfidf_cleaned.head())
print("üìê Taille TF-IDF nettoy√©e :", tfidf_cleaned.shape)

# ‚ûï 6. Features suppl√©mentaires (has_url, etc.)
extra_features = df_text[['has_url', 'has_mention', 'has_hashtag', 'has_emoji']].astype(int)
X_extra = scipy.sparse.csr_matrix(extra_features.values)

# üîó 7. Fusion TF-IDF nettoy√©e + features binaires
X_cleaned_sparse = scipy.sparse.csr_matrix(tfidf_cleaned.values)
X_combined = scipy.sparse.hstack([X_cleaned_sparse, X_extra])
y = df_text['label']

# ‚öñÔ∏è 8. R√©√©quilibrage avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_combined, y)

print("\n‚úÖ R√©partition apr√®s SMOTE :")
print(pd.Series(y_resampled).value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, make_scorer
from sklearn.model_selection import cross_val_score, StratifiedKFold

import warnings
warnings.filterwarnings("ignore")

# ‚úÖ Cross-validation uniquement
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# ‚úÖ Mod√®les
scores = {}
models = {
    "Na√Øve Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "SVM (lin√©aire)": SVC(kernel='linear'),
    "KNN": KNeighborsClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "Neural Network (MLP)": MLPClassifier(max_iter=300)
}

# ‚úÖ Scorers personnalis√©s
scorers = {
    "Accuracy": make_scorer(accuracy_score),
    "Precision": make_scorer(precision_score, average='weighted'),
    "Recall": make_scorer(recall_score, average='weighted'),
    "F1 Score": make_scorer(f1_score, average='weighted')
}

# ‚úÖ √âvaluation cross-validation uniquement
for name, model in models.items():
    print(f"\nüîç √âvaluation par cross-validation : {name}")
    model_scores = {}
    for metric_name, scorer in scorers.items():
        scores_cv = cross_val_score(model, X_resampled, y_resampled, cv=cv, scoring=scorer)
        mean = scores_cv.mean()
        std = scores_cv.std()
        model_scores[metric_name] = f"{mean:.2f} ¬± {std:.2f}"
        print(f"{metric_name}: {model_scores[metric_name]}")
    
    scores[name] = model_scores
    print("-" * 60)

# ‚úÖ R√©sum√© des performances
df_scores = pd.DataFrame(scores).T

# üìã Affichage du tableau
display(df_scores)

# ‚úÖ Visualisation des scores moyens (sans ¬±)
plot_df = df_scores.copy()
for metric in plot_df.columns:
    plot_df[metric] = plot_df[metric].str.extract(r'([0-9.]+)').astype(float)

plot_df[["Accuracy", "Precision", "Recall", "F1 Score"]].plot(kind='bar', figsize=(12, 6))
plt.title("üìä Performances des mod√®les (5-fold CV, SMOTE)")
plt.xlabel("Mod√®le")
plt.ylabel("Score moyen")
plt.xticks(rotation=45)
plt.legend(loc="lower right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# üì¶ Imports n√©cessaires
import pandas as pd
import numpy as np
import re
import emoji
import spacy
import scipy.sparse
from langdetect import detect
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# ‚úÖ Chargement des mod√®les spaCy
nlp_fr = spacy.load("fr_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# ‚úÖ Chargement des stopwords fran√ßais
stopwords_df = pd.read_csv("StopWordsFrench.csv", sep=',', index_col=0)
french_stopwords = stopwords_df.index.tolist()

# ‚úÖ Fonction de nettoyage de texte + lemmatisation multilangue
def clean_text_for_model(text):
    text = text.lower()
    text = re.sub(r"http\S+", "URL", text)
    text = re.sub(r"@\w+", "MENTION", text)
    text = re.sub(r"#(\w+)", r"\1", text)
    text = re.sub(r"[^\w\s:]", "", text)
    return text

def clean_and_lemmatize_multilang(text):
    text = clean_text_for_model(text)
    try:
        lang = detect(text)
    except:
        lang = "en"
    doc = nlp_fr(text) if lang == "fr" else nlp_en(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])

# ‚úÖ Appliquer nettoyage + lemmatisation
print("‚è≥ Application du pr√©traitement avanc√©...")
df_text['text_clean'] = df_text['tweet_text'].apply(clean_and_lemmatize_multilang)
print("‚úÖ Texte nettoy√© et lemmatis√©.")

# ‚úÖ TF-IDF avanc√©
vectorizer = TfidfVectorizer(
    min_df=5,
    max_df=0.95,
    stop_words=french_stopwords,
    lowercase=False,
    ngram_range=(1, 2),
    max_features=50000
)
X_tfidf = vectorizer.fit_transform(df_text['text_clean'])

# ‚úÖ Nettoyage des tokens (filtrage)
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer.get_feature_names_out())

def is_valid_token(token):
    return bool(re.match(r'^[a-zA-Z]+$', token)) and not token.isdigit()

valid_columns = [t for t in tfidf_df.columns if is_valid_token(t)]
numeric_columns = [t for t in tfidf_df.columns if t.isdigit()]
tfidf_df['numeric'] = tfidf_df[numeric_columns].sum(axis=1) if numeric_columns else 0
tfidf_df.drop(columns=numeric_columns, inplace=True)

# ‚úÖ S√©lection finale du TF-IDF filtr√©
tfidf_filtered = tfidf_df[valid_columns + ['numeric']]

# ‚úÖ Ajout des features binaires (d√©j√† calcul√©es)
extra_features = df_text[['has_url', 'has_mention', 'has_hashtag', 'has_emoji']].astype(int)
X_extra = scipy.sparse.csr_matrix(extra_features.values)

# ‚úÖ Fusion TF-IDF filtr√© + features binaires
X_cleaned = scipy.sparse.csr_matrix(tfidf_filtered.values)
X_combined = scipy.sparse.hstack([X_cleaned, X_extra])
y_cleaned = df_text['label']

print("üìê Taille finale avec pr√©traitement :", X_combined.shape)

# ‚úÖ R√©√©quilibrage SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_combined, y_cleaned)

print("‚úÖ R√©partition apr√®s SMOTE :")
print(pd.Series(y_resampled).value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, make_scorer
from sklearn.model_selection import cross_val_score, StratifiedKFold

# ‚úÖ 1. Cross-validation stratifi√©e (sur tout le jeu SMOTE)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# ‚úÖ 2. Liste des mod√®les
scores_v2 = {}
models = {
    "Na√Øve Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "SVM (lin√©aire)": SVC(kernel='linear'),
    "KNN": KNeighborsClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "Neural Network (MLP)": MLPClassifier(max_iter=300)
}

# ‚úÖ 3. D√©finir les scorers personnalis√©s
scorers = {
    "Accuracy": make_scorer(accuracy_score),
    "Precision": make_scorer(precision_score, average='weighted'),
    "Recall": make_scorer(recall_score, average='weighted'),
    "F1 Score": make_scorer(f1_score, average='weighted')
}

# ‚úÖ 4. √âvaluation cross-validation
for name, model in models.items():
    print(f"\nüîç Cross-validation sur : {name}")
    cv_results = {}

    for metric_name, scorer in scorers.items():
        cv_scores = cross_val_score(model, X_resampled, y_resampled, cv=cv, scoring=scorer)
        mean = cv_scores.mean()
        std = cv_scores.std()
        cv_results[metric_name] = f"{mean:.2f} ¬± {std:.2f}"
        print(f"{metric_name}: {cv_results[metric_name]}")

    scores_v2[name] = cv_results
    print("-" * 60)

# ‚úÖ 5. R√©sum√© sous forme de DataFrame
df_scores_v2 = pd.DataFrame(scores_v2).T

# ‚úÖ Affichage du tableau des scores
display(df_scores_v2)

# ‚úÖ 6. Visualisation graphique (moyennes uniquement)
# Extraire juste les moyennes (en float) pour plotting
plot_df_v2 = df_scores_v2.copy()
for metric in plot_df_v2.columns:
    plot_df_v2[metric] = plot_df_v2[metric].str.extract(r'([0-9.]+)').astype(float)

plot_df_v2[["Accuracy", "Precision", "Recall", "F1 Score"]].plot(kind='bar', figsize=(12, 6))
plt.title("üìä Performances des mod√®les (cross-validation sur donn√©es SMOTE)")
plt.xlabel("Mod√®le")
plt.ylabel("Score moyen")
plt.xticks(rotation=45)
plt.legend(loc="lower right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
